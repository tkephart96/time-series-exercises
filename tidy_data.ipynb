Attendance Data

- Read the data from the attendance table and calculate an attendance percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

You should end up with something like this:

name
Billy    0.5250
Jane     0.6875
John     0.9125
Sally    0.7625
Name: grade, dtype: float64


```python
att = pd.read_csv(attendance,index_col=0).reset_index().rename(columns={'index':'name'})
att
```

```python
att = att.melt(id_vars='name')
att.head()
```

```python
att.value = att.value.map({'P':0,'T':.1,'H':.5,'A':1})
att.head()
```

```python
1-att.groupby('name').value.mean()
```

Cake Recipes

- Read the cake_recipes table. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.
- Tidy the data as necessary.
- Which recipe, on average, is the best? recipe b
- Which oven temperature, on average, produces the best results? 275

```python
cr = pd.read_csv(cake)
cr
```

```python
cr[['recipe','position']] = cr['recipe:position'].str.split(':',expand=True)
cr = cr.drop(columns='recipe:position')
cr
```

```python
cr = cr.melt(id_vars=['recipe','position'])
cr
```

```python
cr.groupby('recipe').value.mean()
```
